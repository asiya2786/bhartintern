```python
counter = 0
rs,cs = data.shape

data.drop_duplicates(inplace=True)

if data.shape==(rs,cs):
    print('\n\033[1mInference:\033[0m The dataset doesn\'t have any duplicates')
else:
    print(f'\n\033[1mInference:\033[0m Number of duplicates dropped/fixed ---> {rs-data.shape[0]}')
```

```python
from sklearn.preprocessing import LabelEncoder
categ = ["mainroad","guestroom","basement","hotwaterheating","airconditioning","prefarea","furnishingstatus"]
le = LabelEncoder()
data[categ] = data[categ].apply(le.fit_transform)
```

```python
data.head()
```

```python
corr = data.corr()
plt.figure(figsize=(14,5))
sns.heatmap(corr,cmap='coolwarm',annot=True)
```

```python
X = data.drop(['price'],axis=1)
y = data['price']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
len(X_train),len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
import sklearn.metrics

mse = sklearn.metrics.mean_squared_error(y_test, y_predict)
print('Mean Squared Error on Test Data: {mse}')
```

```python
print('Predicted values:', y_predict)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
score = r2_score(y_test,y_predict)
mae = mean_absolute_error(y_test,y_predict)
```
